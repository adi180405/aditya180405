üìò Project Description

This project aims to predict the selling price of a house based on various features such as its area, number of bedrooms, bathrooms, and amenities. Using a machine learning regression model, the system learns patterns from historical housing data and estimates property prices for new inputs.

The project integrates a Flask web application that allows users to input house details through a simple form and instantly view the predicted price.

1. Objective

To build a predictive model that estimates house prices based on multiple independent variables influencing property value.

The model takes the following features as inputs:

Feature Name	Description	Type

9) parking	Number of parking spots available	Numerical

10) prefarea	Whether the house is located in a preferred area (1 = Yes, 0 = No)	Categorical (Binary)
