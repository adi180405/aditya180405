```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv("employee data.csv")
```

```python
df.head()
```

```python
df.drop(columns=['Employee_ID','Hire_Date','Years_At_Company','Team_Size','Resigned','Training_Hours'],inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
#feature encoding 
from sklearn.preprocessing import LabelEncoder
le1=LabelEncoder()
le2=LabelEncoder()
le3=LabelEncoder()
le4=LabelEncoder()

```

```python
df['Department']=le1.fit_transform(df['Department'])
df['Gender']=le2.fit_transform(df['Gender'])
df['Job_Title']=le3.fit_transform(df['Job_Title'])
df['Education_Level']=le4.fit_transform(df['Education_Level'])
```

```python
df.head()
```

```python
df['Department'].unique()
```

```python
df.info()
```

```python
df.drop(columns=['Job_title'],inplace=True)
```

```python
df.info()
```

```python
x=df[['Department', 'Gender', 'Age', 'Job_Title', 'Education_Level',
     'Monthly_Salary', 'Work_Hours_Per_Week',
       'Projects_Handled', 'Overtime_Hours', 'Sick_Days',
       'Remote_Work_Frequency', 'Promotions', 'Employee_Satisfaction_Score']]
```

```python
y=df['Performance_Score']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
import pickle 
pickle.dump(scaler,open('Standard.pkl','wb'))
```

```python
from sklearn.linear_model import LinearRegression
liner=LinearRegression()
liner.fit(x_train,y_train)
```

```python
y_ped=liner.predict(x_test)
```

```python
from sklearn.metrics import r2_score
print(r2_score(y_test,y_ped))
```

```python
from sklearn.linear_model import LassoCV
lasso=LassoCV(cv=5)
lasso.fit(x_train,y_train)
y_p=lasso.predict(x_test)
```

```python
from sklearn.metrics import r2_score
print(r2_score(y_test,y_p))
```

```python
from sklearn.ensemble import RandomForestRegressor
random=RandomForestRegressor()
random.fit(x_train,y_train)
y_ran=random.predict(x_test)
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(y_test, y_ran)
print("RÂ² Score (Accuracy):", r2)
```

```python
import pickle 
pickle.dump(random,open('random.pkl','wb'))
```
